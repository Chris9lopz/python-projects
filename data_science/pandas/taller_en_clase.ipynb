```python
# Importar librerias
import pandas as pd
import numpy as np
```

```python
# Importar dataframe y ajuste de columnas
df = pd.read_excel("Listado_proveedores.xlsx", header=0, sheet_name= "proveedores", index_col="ID", parse_dates= ["Fecha de última compra"])
# Declaración de columnas
columns = ['PROVEEDOR', 'CONTACTO_COMERCIAL', 'EMAIL', 'TELEFONO', 'SALDO_PENDIENTE', 'FECHA_ULTIMA_COMPRA']
# Cambio de columnas
df.columns = columns
# Información de columnas
df.head()
```

```python
# Cambio de tipo de datos
df = df.astype({
    'PROVEEDOR': "string",
    'EMAIL': "string",
    'CONTACTO_COMERCIAL': "string",
    'TELEFONO': "string"
}) 

df.info()
```

```python
# Identificar los valores null
df.isnull().sum()
```

```python
# Eliminación de valores null
df = df.dropna(how="all")
df.isnull().sum()
```
